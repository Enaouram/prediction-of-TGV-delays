# Fusionner les datasets pour ajouter des features

## Ajout des jours fériés

```python
import pandas as pd
import os

# Charger le dataset des jours fériés
script_dir = os.path.abspath('')
relative_path = 'nb_jours_ferie.csv'
csv_path = os.path.join(script_dir, relative_path)
data_jours_ferie = pd.read_csv(csv_path, delimiter=",")

data_jours_ferie.head()
```

```python
# Charger le dataset des retards de trains
relative_path = 'regularite-mensuelle-tgv-aqst.csv'
csv_path = os.path.join(script_dir, relative_path)
data_retards = pd.read_csv(csv_path, delimiter=";")

data_retards['date'] = pd.to_datetime(data_retards['date'])
data_retards['annee'] = data_retards['date'].dt.year
data_retards['mois'] = data_retards['date'].dt.month

data_retards.head()
```

```python
# Fusionner les deux datasets en utilisant la colonne 'date' comme clé de fusion
merged_data_retard_ferie = pd.merge(data_retards, data_jours_ferie, left_on=['annee', 'mois'], right_on=['annee', 'mois'], how='left')

# Remplace les valeurs NaN par 0
merged_data_retard_ferie = merged_data_retard_ferie.fillna(0)
# Le résultat sera un nouveau dataset avec toutes les colonnes des deux datasets
merged_data_retard_ferie.head()
```
